```python
# import PyTorch Library
import torch
from torch.autograd import variable
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# import scikit-learn library
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split

# import pandas library
import pandas as pd
```

```python
# reading a wine dataset
wine = load_wine()
wine
```

```python
# reading explaination of data frame
pd.DataFrame(wine.data, columns=wine.feature_names)
```

```python
# print target
wine.target
```

```python
# make variables of explaination and target
wine_data = wine.data[0:130]
wine_target = wine.target[0:130]
```

```python
#split dataset to trainset and testset
train_X, test_X, train_Y, test_Y = train_test_split(wine_data, wine_target, test_size = 0.2)

# check the number of datasets
print(len(train_X))
print(len(test_X))
```

```python
'''
now, prepare datas to tensor, so we can use PyTorch!
'''

# train data to tensor
train_X = torch.from_numpy(train_X).float()
train_Y = torch.from_numpy(train_Y).long()

# test data to tensor
test_X = torch.from_numpy(test_X).float()
test_Y = torch.from_numpy(test_Y).long()

# check the number of tensors
print(train_X.shape)
print(test_X.shape)
```

```python
# combination of trainSet
train = TensorDataset(train_X, train_Y)

# checking the 1st data
print(train[0])

# split to mini batch
train_loader = DataLoader(train, batch_size=16, shuffle=True)
```

```python
'''
building a neaural network
'''
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(13, 96)
        self.fc2 = nn.Linear(96, 2)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return F.log_softmax(x)

# make instance
model = Net()
        
```

```python
'''
Train model
'''
# Loss function
criterion = nn.CrossEntropyLoss()

# Optimizer
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Device Setting
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

# Training loop
for epoch in range(300):
    total_loss = 0
    for train_X, train_Y in train_loader:
        # Move data to the same device as the model
        train_X, train_Y = train_X.to(device), train_Y.to(device)

        # Gradient Initializing
        optimizer.zero_grad()

        # Forward cal
        output = model(train_X)

        # Loss cal
        loss = criterion(output, train_Y)

        # Backpropagation
        loss.backward()

        # Weight updating
        optimizer.step()

        # Total loss cal
        total_loss += loss.item()

    # Print loss every 50 epochs
    if (epoch + 1) % 50 == 0:
        print(f"Epoch {epoch + 1}, Loss: {total_loss:.4f}")
```

```python
# 테스트 루프
for test_X, test_Y in test_loader:
    test_X, test_Y = test_X.to(device), test_Y.to(device)
    result = torch.max(model(test_X), 1)[1]
    correct = (test_Y == result).sum().item()
    accuracy = correct / test_Y.size(0)
    print(f"Accuracy: {accuracy:.4f}")
    break  # 테스트용으로 첫 번째 배치만 실행
```
